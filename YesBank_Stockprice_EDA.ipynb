## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
data["Date"] = pd.to_datetime(data['Date'],format='%b-%y',errors='coerce')

# Drop the 'Date' column to keep only numerical data
numeric_data = data.drop('Date', axis=1)

# Calculate Q1, Q3, IQR, and outlier thresholds for each feature
q1 = numeric_data.quantile(0.25)
q3 = numeric_data.quantile(0.75)
iqr = q3 - q1  # Compute IQR
lower_bound = q1 - 1.5 * iqr  # Lower bound for outliers
upper_bound = q3 + 1.5 * iqr  # Upper bound for outliers


# validation (price cannot be lesser than or equal to zero)
numeric_data[(numeric_data <= 0).all(axis=1)].sum()
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code

# Create the box plot
plt.figure(figsize=(10, 6))  # Set figure size for better readability
ax = sns.boxplot(data=numeric_data)

# Add title and labels
plt.title("Box Plot of Numerical Features with IQR and Outlier Thresholds", fontsize=14, fontweight='bold')
plt.xlabel("Features", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Annotate IQR values and outlier thresholds on the plot
for i, feature in enumerate(numeric_data.columns):
    # Get Q1, Q3, and outlier thresholds for the feature
    q1_val = q1[feature]
    q3_val = q3[feature]
    lower_val = lower_bound[feature]
    upper_val = upper_bound[feature]

    # Add text annotation for Q1 and Q3
    ax.text(i, q1_val, f'Q1: {q1_val:.2f}', horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='black')
    ax.text(i, q3_val, f'Q3: {q3_val:.2f}', horizontalalignment='center', verticalalignment='top', fontsize=10, color='black')

    # Add text annotation for lower and upper outlier thresholds

    ax.text(i, upper_val, f'Upper: {upper_val:.2f}', horizontalalignment='center', verticalalignment='top', fontsize=10, color='red')

# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

A box plot is chosen to visualize the distribution, IQR (Interquartile Range), and outlier thresholds for numerical stock price data (Open, High, Low, Close). It effectively highlights central tendency, spread, and extreme values, making it ideal for detecting potential outliers and price fluctuations over time.

##### 2. What is/are the insight(s) found from the chart?

- The presence of multiple outliers above the upper thresholds suggests periods of high volatility and unusual price spikes in Yes Bank's stock prices.

#### Chart - 2

````python
# Chart - 2 visualization code
# Create a 2x2 grid of subplots with shared y-axis
fig, ax = plt.subplots(2, 2, sharey=True, layout='tight')

# Lists to store legend handles and labels (to create a single legend later)
handles, labels = [], []

# Loop through each column in the dataset (excluding the 'Date' column)
for i, col in enumerate(columns[1:]):
    row, col_idx = i // 2, i % 2  # Determine subplot position (row, column)

    # Create a distribution plot for the current column
    sns.distplot(data[col], ax=ax[row, col_idx])


    # Add a vertical line for the mean value (red, solid line)
    line1 = ax[row, col_idx].axvline(data[col].mean(), color='red', linewidth=2, label='Mean')

    # Add a vertical line for the median value (black, dashed line)
    line2 = ax[row, col_idx].axvline(data[col].median(), color='black', linestyle='dashed', linewidth=2, label='Median')

    # Set a title for the current subplot, naming it after the column
    ax[row, col_idx].set_title(f"Distribution of {col} Price", fontsize=12)

    # Collect legend handles and labels only once (from the first subplot)
    if i == 0:
        handles.extend([line1, line2])
        labels.extend(['Mean', 'Median'])

# Create a single legend for all subplots using the collected handles and labels
fig.legend(handles, labels, loc='upper right')

# Set a global title for the entire figure
fig.suptitle("Distribution Plots with Mean and Median(capped data)", fontsize=14, fontweight='bold')

# Display the plot
plt.show()
````

A distribution plot (with mean and median) was chosen to analyze the spread and skewness of Yes Bank's stock prices (Open, High, Low, Close). It provides insights into data distribution, central tendency, skewness and outlier impact.

#### Chart - 8

````python
# Chart - 8 visualization code
window_size = 12

# Calculate rolling mean & std
rolling_mean = data["Close"].rolling(window=window_size).mean()
rolling_std = data["Close"].rolling(window=window_size).std()

# Compute Coefficient of Variation (CV)
data["CV"] = rolling_std / rolling_mean

# Plot
plt.figure(figsize=(12, 6))
plt.plot(data["Date"], data["CV"], color='purple')
plt.xlabel("Date")
plt.ylabel(f"Rolling {window_size}-Month CV")
plt.title(f"Coefficient of Variation (CV) Over {window_size} Months", fontsize=14)
plt.grid()
plt.show()
````

##### 1. Why did you pick the specific chart?

The Coefficient of Variation (CV) chart is useful because it measures the relative risk (volatility) compared to the stock’s average price. Unlike absolute volatility, CV helps determine how unstable the stock is relative to its mean price over time.

##### 2. What is/are the insight(s) found from the chart?

- There were periods of stability between 2012 and 2017, where CV remained low, suggesting a more predictable price movement.

````python
# Chart - 11 visualization code
# Define custom SMA and EMA period
sma_period = 6

# Calculate moving averages
data[f'SMA{sma_period}'] = data['Close'].rolling(sma_period).mean()
data[f'EMA{sma_period}'] = data['Close'].ewm(span=sma_period, adjust=False).mean()

# Plotting
plt.figure(figsize=(12,6))
sns.lineplot(x=data["Date"], y=data[f'SMA{sma_period}'], label=f"{sma_period}-Month SMA", color='orange', linestyle="dashed")
sns.lineplot(x=data["Date"], y=data[f'EMA{sma_period}'], label=f"{sma_period}-Month EMA", color='green', linestyle="dotted")
sns.lineplot(x=data["Date"], y=data['Close'], label='Closing Price', color='blue')

# Labels and title
plt.xlabel("Date")
plt.ylabel("Price")
plt.title(f"{sma_period}-Month SMA & EMA of Yes Bank Stock", fontsize=14)
plt.legend()
plt.grid()
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart was chosen because it provides a clear visualization of Yes Bank's stock price movement along with two key technical indicators: 6-Month SMA (Simple Moving Average) and 6-Month EMA (Exponential Moving Average). These indicators help in identifying trend direction, momentum shifts, and potential buy/sell signals, making it an essential tool for traders and investors analyzing historical price patterns.

#### Chart - 12

````python
# Chart - 11 visualization code
# Calculate Short & Long EMA
short_ema = data['Close'].ewm(span=12, adjust=False).mean()  # 12-month EMA
long_ema = data['Close'].ewm(span=26, adjust=False).mean()   # 26-month EMA

# MACD Calculation
macd = short_ema - long_ema  # MACD Line
signal = macd.ewm(span=9, adjust=False).mean()  # Signal Line

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data["Date"], macd, label="MACD", color='blue')
plt.plot(data["Date"], signal, label="Signal Line", color='red')

# Labels and title
plt.title("MACD Indicator", fontsize=14, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend()
plt.grid()
plt.show()
````

##### 1. Why did you pick the specific chart?

The **MACD (Moving Average Convergence Divergence) Indicator** was chosen because it is a widely used **momentum-based technical indicator** that helps in identifying **trend direction, strength, and potential reversal points** in stock prices. This chart specifically highlights the relationship between the **MACD line (blue) and the Signal line (red)**, making it useful for traders and investors to make informed decisions.
